```python
#   3.1  -  Insert records from 42_District_wise_crimes_committed_against_women_2001_2012.csv into a table
```

```python
import sqlite3

import csv
```

```python
db = sqlite3.connect('42_District_wise_crimes_committed_against_women_2001_2012.db')

cursor = db.cursor()
```

```python
cursor.execute('create table Crime_Women(State_UT Text, District Text, Year Int, Rape Int, Kidnapping Int, Dowry Int, Assault Int, Insult Int, Cruelty Int, Importation Of Girls Int ) ')
```

```python
with open('42_District_wise_crimes_committed_against_women_2001_2012.csv','r') as file:
    no_records=0
    
    for row in file:
        cursor.execute('INSERT INTO Crime_Women VALUES(?,?,?,?,?,?,?,?,?,?)',row.split(','))
        
        db.commit()
        no_records+=1
        
        
print('Row Inserted -',no_records)
```
